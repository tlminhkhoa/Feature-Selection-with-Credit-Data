Preprocess Data

```python
X , y = df.iloc[:, df.columns != 'TARGET'], df["TARGET"]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

names = list(df.columns)
names.remove("TARGET")
X_train = pd.DataFrame(X_train,columns =names)
X_train
```

```python
X_train = pd.get_dummies(X_train)
X_train_name = X_train.columns.tolist()
X_train
```

```python
X_train_countUnique= X_train.select_dtypes(exclude = "object").nunique().sort_values()

X_train_countUnique[(X_train_countUnique > 50) & (X_train_countUnique < 1500)]
```

```python
X_train_countUnique[(X_train_countUnique > 50) & (X_train_countUnique < 1500)].plot(rot = 20)
```

```python
numeric_columns = list(X_train_countUnique[X_train_countUnique > 364].index)
```

```python
def _zoom_3sigma(col, dataset):
    xs = dataset[col]
    mu = xs.mean()
    sigma = xs.std()
    low =  mu - 3*sigma
    high = mu + 3*sigma
    
    def _value(x):
        if x < low: return low
        elif x > high: return high
        else: return x
    xnew = list(map(lambda x: _value(x), xs))
  

    return xnew ,mu,low , high

train_dict = {}
for col in numeric_columns:
    X_train[col],mu,low , high = _zoom_3sigma(col, X_train) 
    train_dict[col] = {}
    train_dict[col]["mu"] = mu
    train_dict[col]["low"] = low
    train_dict[col]["high"] = high
```

```python
from sklearn.impute import SimpleImputer
imputerMean = SimpleImputer(strategy='mean')
X_train = imputerMean.fit_transform(X_train)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train
```

Train base model

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import auc
from sklearn.metrics import plot_roc_curve
import matplotlib.pyplot as plt

def cv_auc(clf,X_input, y_input,cv =10):

  aucs = []
  
  
  fig, ax = plt.subplots()
  
  cv = StratifiedKFold(n_splits=10)
  
  for i, (train, test) in enumerate(cv.split(X_input, y_input)):
    print(i)
    clf.fit(X_input[train], y_input[train])
    viz = plot_roc_curve(clf, X_input[test], y_input[test],
                         name='ROC fold {}'.format(i),
                         alpha=0.3, lw=1,ax=ax)
    
    
    aucs.append(viz.roc_auc)
  
  ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
        label='Chance', alpha=.8)
  return aucs
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0,n_jobs = -1)
aucLog = cv_auc(clf,X_train,y_train.to_numpy())
aucLog
```

```python
from sklearn.ensemble import RandomForestClassifier

rd_classifier = RandomForestClassifier(max_depth = 5, n_jobs = -1,random_state = 0)
aucRR = cv_auc(rd_classifier,X_train,y_train.to_numpy())
aucRR
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
aucGNB = cv_auc(gnb,X_train,y_train.to_numpy())
aucGNB
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='adam', random_state=0,early_stopping = True,max_iter = 500)
aucNN = cv_auc(clf,X_train,y_train.to_numpy())
aucNN
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
aucDT = cv_auc(clf,X_train,y_train.to_numpy())
aucDT
```

```python
scores = pd.DataFrame({'aucLog':aucLog,'aucRR':aucRR,'aucLog':aucLog,'aucGNB':aucGNB,'aucNN':aucNN,'aucDT':aucDT})
scores
```

```python
import scipy.stats as stats

#perform the paired samples t-test
pvalue = stats.ttest_rel(scores["aucLog"], scores["aucRR"]).pvalue

pvalue < 0.05
# aucLog > aucRR
```

```python
def _zoom_test(col,X_test):
  xs = X_test[col]
  high = train_dict["{}".format(col)]["high"]
  mu = train_dict["{}".format(col)]["mu"]
  low = train_dict["{}".format(col)]["low"]

  def _value(x):
        if x < low: return low
        elif x > high: return high
        else: return x
  xnew = list(map(lambda x: _value(x), xs))
  

  return xnew

for col in numeric_columns:
    X_test[col] = _zoom_test(col, X_test)
```
